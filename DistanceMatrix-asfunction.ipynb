```julia
# The list of original sequences
sequences = [
"atcgctcacgacaagcatcagcatcagcatcatacgtgactcaagaaGAATAtcatgttaAATGGTTttactaccccctgtggggatatcgatctactactacgact",
"atctagccgatgattctaccacgatctcatactacttagcattagcatcatacgcgatcgggggatatatctctactacgtctcagacacgactccgact",
"atcggtcactacttgcatcagcaTTGCTCAATCTGtcagcatgcatacCGTCCACCATAAGAAAAGATGGtcgaggatcatcttattactacccccgcgggggatatatatcttctacgactcagctaAACATTCAAcgattacgact",
"atcgttctacgagctcatactactagcatcagcatcagcatCCCATTAgatacgcgatcagcacatcttattactacccccgccggggatatatacagcagcagcatcgactctactacgactt",
"atcggtcactactagcatcagatcaccatactcgagcatcatcttattactaccccgcgcatactacgactaccgcggtggatatatatctactaccactcagctacgactacgact"
]

# A list of lists of lists :) -- let's break it down:
#     The top level list is of matches
#     The next level list is the sequences (nothing means there was no match in the sequence)
#     The next list is the information for each sequence with that match:
#         1st index - orientation: forward = 0, reverse complement = 1
#         2nd index - start position of the match
#         3rd index - end position of the match
# Ex. matches[1][2][2] = 33 -- the first match, the second sequence with that match, 
#                              the start position of the match in that sequence
matches = [
[nothing, nothing, nothing, [0, 1, 7], [0, 1, 7]], 
[nothing, nothing, [0, 1, 21], nothing, [0, 1, 21]], 
[nothing, [0, 50, 59], nothing, [0, 50, 59], nothing], 
[[0, 1, 21], nothing, nothing, nothing, [0, 1, 21]], 
[[0, 68, 77], nothing, [0, 86, 95], [0, 71, 80], [0, 47, 56]], 
[[0, 82, 89], [0, 61, 68], [0, 100, 107], [0, 85, 92], nothing], 
[[0, 50, 56], [0, 24, 30], [0, 80, 86], nothing, nothing], 
[nothing, nothing, [0, 10, 21], [0, 21, 32], [0, 10, 21]], 
[[0, 11, 31], nothing, nothing, [0, 22, 42], nothing], 
[nothing, [0, 15, 21], nothing, nothing, [0, 32, 38]], 
[[0, 33, 39], nothing, nothing, [0, 51, 57], nothing], 
[[0, 21, 39], [0, 39, 57], nothing, nothing, nothing], 
[nothing, [0, 69, 75], nothing, [0, 89, 95], nothing], 
[nothing, [0, 28, 35], nothing, [0, 16, 23], [0, 60, 67]], 
[[0, 68, 78], nothing, [0, 86, 96], [0, 71, 81], nothing], 
[nothing, [0, 87, 99], nothing, [0, 110, 122], nothing], 
[nothing, [0, 68, 74], nothing, nothing, [0, 94, 100]], 
[[0, 26, 33], [0, 44, 51], nothing, nothing, [0, 35, 42]], 
[nothing, [0, 89, 95], [0, 117, 123], nothing, [0, 95, 101]], 
[nothing, [0, 59, 85], [0, 98, 124], nothing, nothing], 
[[0, 10, 16], nothing, nothing, [0, 49, 55], nothing], 
[[0, 8, 14], nothing, [0, 54, 60], nothing, nothing], 
[nothing, nothing, nothing, [0, 48, 54], [0, 102, 108]], 
[[0, 28, 34], [0, 46, 52], nothing, [0, 64, 70], nothing], 
[nothing, [0, 60, 79], [0, 99, 118], nothing, [0, 77, 96]], 
[nothing, [0, 65, 72], [0, 106, 113], nothing, [0, 84, 91]], 
[nothing, [0, 91, 97], [0, 139, 145], nothing, [0, 108, 114]], 
[[0, 64, 70], [0, 63, 69], [0, 102, 108], [0, 87, 93], [0, 80, 86]], 
[nothing, [0, 35, 41], [0, 126, 132], nothing, nothing], 
[nothing, [0, 12, 24], nothing, [0, 109, 121], nothing], 
[[0, 60, 66], [1, 31, 37], [1, 8, 14], [1, 118, 124], nothing], 
[nothing, [0, 33, 39], [1, 82, 88], [1, 67, 73], [1, 43, 49]], 
[nothing, [0, 4, 10], [1, 122, 128], nothing, [1, 100, 106]], 
[nothing, nothing, [0, 110, 116], [1, 60, 66], [0, 88, 94]], 
[[0, 58, 67], nothing, [1, 128, 137], nothing, nothing], 
[nothing, nothing, [0, 132, 138], [0, 17, 23], [0, 61, 67]], 
[nothing, nothing, nothing, [0, 91, 97], [1, 79, 85]], 
[nothing, nothing, nothing, [0, 98, 104], [1, 70, 76]], 
[[0, 72, 88], nothing, nothing, [1, 75, 91], nothing], 
[nothing, nothing, [0, 77, 83], [1, 12, 18], [0, 38, 44]], 
[nothing, [0, 50, 56], [1, 72, 78], nothing, nothing], 
[nothing, [0, 70, 76], [0, 6, 12], nothing, [0, 6, 12]], 
[[0, 35, 41], [1, 85, 91], nothing, nothing, nothing], 
[nothing, [0, 5, 17], nothing, [1, 100, 112], nothing], 
[[0, 85, 91], [1, 22, 28], nothing, nothing, nothing], 
[nothing, nothing, [0, 113, 120], [0, 5, 12], nothing], 
[nothing, [0, 21, 27], [1, 65, 71], [0, 9, 15], nothing], 
[nothing, nothing, [0, 38, 44], nothing, [1, 59, 65]], 
[nothing, [0, 71, 77], [1, 60, 66], [0, 111, 117], nothing], 
[[0, 88, 96], nothing, nothing, [0, 1, 9], nothing], 
[nothing, [0, 83, 89], nothing, nothing, [0, 2, 8]], 
[[0, 97, 105], [0, 16, 24], nothing, [0, 113, 121], nothing], 
[[0, 89, 95], nothing, [0, 28, 34], nothing, nothing], 
[[0, 95, 101], nothing, [0, 84, 90], nothing, [0, 45, 51]], 
[[0, 42, 48], nothing, [1, 9, 15], nothing, nothing], 
[nothing, nothing, [0, 79, 95], [0, 64, 80], [0, 40, 56]], 
[nothing, [0, 43, 50], nothing, [0, 24, 31], [0, 13, 20]], 
[nothing, [0, 39, 45], [0, 15, 21], nothing, [0, 15, 21]], 
[[0, 98, 107], [0, 74, 83], nothing, [0, 114, 123], [0, 62, 71]], 
[nothing, [0, 41, 47], [0, 17, 23], nothing, [0, 22, 28]], 
[[0, 96, 102], nothing, [0, 137, 143], [0, 19, 25], nothing], 
[[0, 56, 62], nothing, [0, 25, 31], nothing, nothing], 
[nothing, [0, 6, 12], [1, 16, 22], nothing, nothing], 
[[0, 13, 22], [1, 6, 15], nothing, nothing, nothing], 
[nothing, [0, 86, 92], [0, 61, 67], nothing, nothing], 
[nothing, nothing, [0, 97, 128], nothing, [0, 75, 106]], 
[[0, 9, 15], [1, 3, 9], nothing, nothing, nothing], 
[[0, 68, 77], nothing, nothing, [0, 71, 80], [0, 47, 56]], 
[nothing, nothing, [0, 79, 110], [0, 64, 95], nothing], 
[nothing, nothing, [0, 55, 61], [0, 43, 49], nothing], 
[nothing, [0, 21, 36], nothing, [0, 9, 24], nothing], 
[[0, 39, 45], nothing, nothing, nothing, [0, 30, 36]], 
[[0, 97, 107], [0, 73, 83], nothing, [0, 113, 123], nothing], 
[nothing, nothing, [0, 43, 49], [0, 38, 44], nothing], 
[nothing, nothing, nothing, [0, 105, 111], [0, 101, 107]], 
[[0, 93, 100], [0, 72, 79], nothing, [0, 112, 119], nothing]
]




# Sorts matches to be longest to shortest
function sortMatches(matchA::Array{Union{Nothing, Array{Int64,1}},1}, matchB::Array{Union{Nothing, Array{Int64,1}},1})
    
    firstA = findfirst(i->i!=nothing, matchA)
    firstB = findfirst(i->i!=nothing, matchB)
    lenA = matchA[firstA][3] - matchA[firstA][2]
    lenB = matchB[firstB][3] - matchB[firstB][2]

    return(lenA < lenB)

end

# Returns matches that exist in both pairwise sequences
function keepMatches(i::Int64, j::Int64, matches::Array{Array{Union{Nothing, Array{Int64,1}},1},1})
    
    cur_matches = Array{Union{Nothing, Array{Int64,1}},1}[]
    for match in matches
        if match[i]!=nothing && match[j]!=nothing
            push!(cur_matches, [deepcopy(match[i]), deepcopy(match[j])])
        end
    end

    return(cur_matches)
end


function trimMatches(a::Array{Union{Nothing, Array{Int64,1}},1}, b::Array{Union{Nothing, Array{Int64,1}},1}) 

    #print("type: ", a, "\n")

    # Checking if the second match overlapped with the first match in either sequence
    while ( b[1][2] <= a[1][3] && b[1][2] >= a[1][2] ) || ( b[2][2] <= a[1][3] && b[2][2] >= a[1][3] )
        # move the start position up until it no longer overlaps with the other match
        b[1][2] += 1
        b[2][2] += 1
    end
    
    while ( b[1][3] <= a[1][3] && b[1][3] >= a[1][2] ) || ( b[2][3] <= a[1][3] && b[2][3] >= a[1][3] )
        # move the end position down until it no longer overlaps with the other match
        b[1][3] -= 1
        b[2][3] -= 1
    end
    
    if b[1][3] - b[1][2] <= 0 # a match has been completely removed
        b[1] = nothing
        b[2] = nothing
    end
end


# Calculates coverage -- can be weighted on which nucleotide has been substituted
function ntWeight(g1::String, g2::String, matches::Array{Array{Union{Nothing, Array{Int64,1}},1},1})
    
    coverageA = 0
    coverageB = 0
    
    strong = .98
    weak = .99
    purine_transition = 1.1
    pyrimidine_transition = 1.1
    transversion = 1.05
    
    for m in 1:length(matches)
        if matches[m][1] != nothing
            ntSeq2 = matches[m][2][2] # starting point nucleotide in the second sequence
            for ntSeq1 in matches[m][1][2]:matches[m][1][3] # starting point nucleotide in the first sequence

                if (g1[ntSeq1]=="a" && g2[ntSeq2]=="a") || (g1[ntSeq1]=="t" && g2[ntSeq2]=="t") # strong
                    coverageA += strong
                    coverageB += strong
                elseif (g1[ntSeq1]=="g" && g2[ntSeq2]=="g") || (g1[ntSeq1]=="c" && g2[ntSeq2]=="c") # weak
                    coverageA += weak
                    coverageB += weak
                elseif (g1[ntSeq1]=="a" && g2[ntSeq2]=="g") || (g1[ntSeq1]=="g" && g2[ntSeq2]=="a") # purine
                    coverageA += purine_transition
                    coverageB += purine_transition
                elseif (g1[ntSeq1]=="c" && g2[ntSeq2]=="t") || (g1[ntSeq1]=="t" && g2[ntSeq2]=="c") #pyrimidine
                    coverageA += pyrimidine_transition
                    coverageB += pyrimidine_transition
                else # transversion substitution
                    coverageA += transversion
                    coverageB += transversion
                end
                
                ntSeq2 += 1
                
            end
        end
    end
    
    return coverageA, coverageB
    
end


########## Begin Processing Matches to Find Edit Distances ##########

function findDistances(sequences, matches)
    
    # Make sure that the incoming sequences are all lowercase
    seqnum=1
    for seq in sequences
        sequences[seqnum] = lowercase(seq)
        seqnum += 1
    end
    
    #initialize output matrix
    edit_distance_matrix = zeros(Float64, (length(sequences), length(sequences))) 


    # Step 1) Sort the matches by descending length 
    sort!(matches, lt=sortMatches, rev=true)

    # Pairwise Comparison
    for i in 1:length(sequences)                
        for j in (i+1):length(sequences)

            # Step 2) Subset the matches array to only contain matches that exist 
            #         for the current pair of sequences being compared
            cur_matches = keepMatches(i,j,matches)

            for k in 1:length(cur_matches)
                if cur_matches[k][1] == nothing # for the case that an entire match was trimmed away in a previous round
                    continue
                end
                for l in (k+1):length(cur_matches)
                    if cur_matches[l][1] == nothing # for the case that an entire match was trimmed away
                        continue
                    end

                    # Step 3) Trim overlapping matches in each sequence
                    trimMatches(cur_matches[k], cur_matches[l])

                end
            end

            # Step 4) Calculate the coverage:
            #         Coverage within a match can be weighted based on the type of match/substitiution
            #         Nucleotides not in a match are counted as dissimilar.
            coverage = ntWeight(sequences[i], sequences[j], cur_matches)
            coverageA = coverage[1]
            coverageB = coverage[2]

            percentA = coverageA / length(sequences[i])
            percentB = coverageB / length(sequences[j])
            percentAB = (percentA + percentB) / 2   # average the coverage
            editPercent = 1 - percentAB 

            # Step 5) Set the edit distance for the pairwise genomes
            edit_distance_matrix[i, j] = editPercent

            # Step 6) Repeat until all pairs have a distance calculated for them
        end
    end
    
    return edit_distance_matrix
end

findDistances(sequences, matches)
```
